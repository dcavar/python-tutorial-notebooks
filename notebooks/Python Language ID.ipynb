```python
from collections import Counter
```

```python
enTfp = Counter(enTrigrams + [ ';;;' ])
#print(enTfp)
totalEn = sum(enTfp.values())
enProbs = dict( [ (x[0], x[1]/totalEn) for x in enTfp.items() ] )
print(enProbs)
```
